## Testing Loaded Models

```python
import joblib
```

```python
df_test = pd.read_json('../raw_data/data_2500_labeled.json')[:50]
df_test.head()
```

```python
from classipy import CustomLabelEncoder
```

```python
label_encoder = CustomLabelEncoder()
```

```python
X = JSONtoModelTransformer().fit_transform(df_test)
y = df_test['label']

model = joblib.load('../classipy/models//trained_models/RandomForest.joblib')
y_preds = model.predict(X)

comparision_df = pd.DataFrame({
                    'y_preds':y_preds,
                    'decoded_y_preds': label_encoder.decode_labels(y_preds),
                    'actual_labels': y
                })
comparision_df.head()
```
